```python
import pandas as pd
from sklearn.decomposition import PCA,FactorAnalysis
from sklearn.preprocessing import StandardScaler
import numpy as np
import seaborn as sns
from factor_analyzer.factor_analyzer import calculate_kmo,calculate_bartlett_sphericity
from factor_analyzer import FactorAnalyzer
```

```python
water = pd.read_csv('Water_pond_tanks_2021.csv',encoding='latin1')
water
```

```python
for column in water.columns:
  water.loc[water[column]=='-',column] = np.nan
  water.loc[water[column]=='7.54\n4',column] = 7.54
water
```

```python
water.iloc[:,4:water.shape[1]+1]=water.iloc[:,4:water.shape[1]+1].astype(float)
water.dtypes
```

```python
def water_quality_index():
  variables = {'pH (Max)':9,'Dissolved Oxygen (mg/L) (Max)':4,'BOD (mg/L) (Max)':3,'Nitrate N + Nitrite N(mg/L) (Max)':50,
               'Total Coliform (MPN/100ml) (Max)':5000,'pH (Min)':9,'Dissolved Oxygen (mg/L) (Min)':4,'BOD (mg/L) (Min)':3,'Nitrate N + Nitrite N(mg/L) (Min)':50,
               'Total Coliform (MPN/100ml) (Min)':5000}
  K = 1/sum(1/np.array(list(variables.values())))
  W = []
  V_ideal = [7,14,0,0,0,7,14,0,0,0]
  WQI = []
  for key in variables.keys():
    W.append(K/variables[key]) 
  for j in range(water.shape[0]):
    row = water.iloc[j,]
    WQ = []
    for i,key in enumerate(variables.keys()):
      WQ.append(W[i]*100*((row[key]-V_ideal[i])/(variables[key]-V_ideal[i])))
    WQI.append(sum(WQ))
  return(WQI)
water['WQI'] = water_quality_index()

water['quality'] = pd.cut(water['WQI'],[0,50,100,200,300,np.inf],labels=['Excellent','Good','Poor','Very Poor','Unfit'])
#water['quality'] = pd.cut(water['WQI'],[0,100,np.inf],labels=['Good','Poor'])
```

```python
X = water.drop(['STN\nCode','Name of Monitoring Location','Type Water Body','State Name','WQI','quality'],axis=1)
X
```

```python
Z = StandardScaler().fit_transform(X[X.notna().all(axis=1)])
```

```python
pca = PCA().fit(Z)
comp = pca.transform(Z)
```

```python
pca.explained_variance_
```

```python
fa = FactorAnalyzer(6,rotation='varimax',method='principal')
fa.fit(Z)

fa_comp = fa.transform(Z)
fa_comp = pd.DataFrame(fa_comp)
L = pd.DataFrame(fa.loadings_,index=X.columns,columns = list(range(1,7))) 
```

```python
np.diag(np.array(pd.DataFrame(Z).cov()) - np.dot(fa.loadings_,fa.loadings_.T)) 
```
